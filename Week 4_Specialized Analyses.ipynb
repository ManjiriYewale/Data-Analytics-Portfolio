### 1️⃣ Daily COVID Case Trend

```python
covid_df = pd.read_csv("complete.csv")

daily_cases = covid_df.groupby('Date')['New cases'].sum()

plt.figure()
daily_cases.plot()
plt.title("Daily COVID Cases Trend")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.show()
plt.tight_layout()
```

### 2️⃣ Death Rate Analysis

```python
import pandas as pd
import numpy as np

# Convert columns to numeric (fix string issue)
covid_df['Death'] = pd.to_numeric(covid_df['Death'], errors='coerce')
covid_df['New cases'] = pd.to_numeric(covid_df['New cases'], errors='coerce')

# Replace 0 in New cases to avoid division error
covid_df['New cases'].replace(0, np.nan, inplace=True)

# Calculate Death Rate
covid_df['Death_Rate'] = (covid_df['Death'] / covid_df['New cases']) * 100

# Calculate average death rate
avg_death_rate = covid_df['Death_Rate'].mean()

print("Average Death Rate:", avg_death_rate)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(covid_df['Date'], covid_df['Death_Rate'])

plt.title("COVID Death Rate Trend")
plt.xlabel("Date")
plt.ylabel("Death Rate (%)")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### 3️⃣ Monthly Trend Analysis

```python
monthly_cases = covid_df.groupby('Date')['New cases'].sum()

plt.figure(figsize=(12,6))
monthly_cases.plot()

plt.title("Monthly COVID Cases")
plt.xlabel("Month")
plt.ylabel("Cases")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Project 5- Finance Analysis (Stock Market Trends)

```python
df=pd.read_csv("tesla-stock-price.csv")
```

```python
df
```

```python
df.head()
```

### Data Cleaning

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['date'])

# Sort by Date
df = df.sort_values('date')

# Check missing values
print(df.isnull().sum())
```

### Stock Price Trend Analysis

```python
plt.figure()
plt.plot(df['Date'], df['close'])
plt.title("Tesla Stock Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```

### Daily Returns Calculation

```python
df['Daily_Return'] = df['close'].pct_change() * 100

print(df[['Date', 'close', 'Daily_Return']].head())
```

### Volatility Calculation

```python
volatility = df['Daily_Return'].std()
print("Stock Volatility (Std Dev of Returns):", volatility)
```

### 7-Day & 30-Day Moving Average

```python
df['MA_7'] = df['close'].rolling(window=7).mean()
df['MA_30'] = df['close'].rolling(window=30).mean()

plt.figure()
plt.plot(df['Date'], df['close'])
plt.plot(df['Date'], df['MA_7'])
plt.plot(df['Date'], df['MA_30'])
plt.title("Tesla Stock with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

### Monthly Average Closing Price

```python
df['Month'] = df['Date'].dt.to_period('M')

monthly_avg = df.groupby('Month')['close'].mean()

plt.figure()
monthly_avg.plot()
plt.title("Monthly Average Closing Price")
plt.xlabel("Month")
plt.ylabel("Average Price")
plt.show()
```

### Highest & Lowest Price Analysis

```python
max_price = df['close'].max()
min_price = df['close'].min()

print("Highest Closing Price:", max_price)
print("Lowest Closing Price:", min_price)
```
